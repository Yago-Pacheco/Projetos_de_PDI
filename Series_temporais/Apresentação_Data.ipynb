## Formatos & Conversões



*   **`date`:** O objeto date representa uma data (ano, mês e dia) em um calendário idealizado, o atual calendário Gregoriano estendido indefinidamente em ambas as direções. Um exemplo de como chamar `datetime.date(year, month, day)`. Todos os argumento são obrigatórios. Dentro do `date` temos outras funções que são chamadas a partir dele.
*   **`time`:** Um objeto `time` representa a hora (local) do dia, independente de qualquer dia em particular, e sujeito a ajustes através de um objeto `tzinfo` (explicando de forma informal é um subclasse que ajuda na construção de objetos). Um exemplo de como chamar é `datetime.time(hour=0, minute=0, second=0, microsecond=0, tzinfo=None, *, fold=0)`. Todos os argumentos são opcionais.
*   **`datetime`:** Um objeto `datetime` é um único objeto contendo todas as informações de um objeto `date` e um objeto `time`. Assim como um objeto `date`, `datetime` assume o atual calendário Gregoriano estendido em ambas as direções; assim como um objeto `time`, `datetime` assume que existem exatamente 3600*24 segundos em cada dia. Um exemplo de como chamar é `atetime.datetime(year, month, day, hour=0, minute=0, second=0, microsecond=0, tzinfo=None, *, fold=0)`. Os argumentos year, month e day são obrigatórios.


**Biblioteca Datetime**

Como mencionado temos algumas diferentes formas de representar as datas sendo a principal mudança a ordem. Aqui será apresentando as bibliotecas que tratam esse formato de data. Claro, pode ser usar a data como um string mas não é o melhor caminho para certas situações. Uma ressalva é que dentro do Python tempos uma biblioteca nativa que é o `Datetime`.

Não só podemos configurar as datas como também podemos fazer cálculos com elas e assim criar medidas ou apenas se informar com o que se pretende com essa operação matemática. ## **Séries Temporais para análise de Dados**

##**Regressão Linear**

A regressão linear é uma técnica de análise de dados que prevê o valor de dados desconhecidos usando outro valor de dados relacionado e conhecido. Ele modela matematicamente a variável desconhecida ou dependente e a variável conhecida ou independente como uma equação linear. Por exemplo, suponha que você tenha dados sobre suas despesas e receitas do ano passado. As técnicas de regressão linear analisam esses dados e determinam que suas despesas são metade de sua renda. Eles então calculam uma despesa futura desconhecida reduzindo pela metade uma renda futura conhecida.

**Relação Linear**

Para que se use a regressão linear é necessário que se tenha uma relação linear entre as variáveis independentes e dependentes. Uma forma simples de reproduzir isso é plotar um gráfico de dispersão (scatter plot) e verificar se eles se encaixam em uma linha reta ( / ou \ ). Caso isso não ocorrer, podemos aplicar algumas tranformações aos dados, podemos aplicar funções não lineares nos dados como raiz quadrada ou logarítmica, para assim criar matematicamente a relação linear entre duas variáveis.

**Independência residual**

Fazemos o uso dos resíduos para medir a recisão da previsão. O resíduo é a diferença entre os dados observados e o valor previsto. Vale resslatar que eles não devem ter um padrão identificável entre eles. Por exemplo, resíduos nãoi devem crescer ao longo do tempo ou apenas diminuir ao longo do tempo.

**Normalidade**

São técnicas de representação gráficas para verificar se os resíduos são normalmente distribuídos. Os resíduos devem se encaixar ao longo de uma linha diagonal no centro do gráfico. Caso não sejam normalizados pode se fazer a verificação da presença de outliers aleatórios ou valores atípicos.

**Exemplo**

Abaixo deixei plotado um exemplo de regressão linear simples, a alguns conceitos como os coeficientes, R², etc. Que serão abordados quando formos falar de predição e os algoritimos usados para estes fins e dentro deles temos o de regressão linear.

```python
# calculando a reta
Y = df_st['Close']
x = np.arange(1,367,1)
n = np.size(df_st['Close'])

data = df_st['Date']

# coeficiente angular
a = (n * np.sum(x*Y) - np.sum(x)*np.sum(Y)) / (n * np.sum(x**2) - np.sum(x)**2)

# costante
b = np.mean(Y) - a * np.mean(x)

# reta
y = a*x+b

# plotando
plt.plot(data, Y)
plt.plot(data, y)
```

Muitas vezes utilizamos a regressão apenas como uma forma de analisar uma tendência, se ela tende a crescer ou diminuir. Porém, assim como na predição ela tem seus limitantes e afim de se fazer uma melhor analise temos algumas alterantivas como as `médias móveis`, `correlação serial` e a `autocorrelação`.

###**Médias Móveis**

Quando nos deparamos com uma tendência que não é uma simples função, uma boa alternativa é fazer uso da média móvel. De forma geral ela divide a tela em regiões sobreposta que chamamos de janelas e calcula a média dos valores de cada janela. Temos udiversos usos em análises para as médias móveis como:

* Para identificar tendências;
* Para identificar pontos de entrada e saída;

Algo importante a se ressaltar as médias móveis tem a sua limitação quando se tem algo muito volátil, de tendência forte ou uma baixa liquidez. Esses exemplos são mais aplicados ao mercado.

###**Correlação Serial**

Correlação serial é a correlação entre os valores de uma série temporal em diferentes períodos de tempo. Ela é uma medida da dependência entre os valores da série temporal.

A correlação serial pode ser positiva, negativa ou nula. Uma correlação serial positiva indica que os valores da série temporal tendem a se mover na mesma direção. Uma correlação serial negativa indica que os valores da série temporal tendem a se mover em direções opostas. Uma correlação serial nula indica que não há relação entre os valores da série temporal.

A correlação pode ser influenciada por algusn fatores como:

* Tendências: Se a série temporal estiver em tendência, os valores da série temporal tenderão a se mover na mesma direção.
* Sazonalidade: Se a série temporal apresentar sazonalidade, os valores da série temporal tenderão a se mover de forma semelhante em determinados períodos do ano.
* Efeitos de memória: Se os valores da série temporal forem influenciados por eventos passados, a correlação serial será positiva.
* Erros de medição: Se os dados da série temporal estiverem contaminados por erros de medição, a correlação serial pode ser positiva ou negativa.

Ainda dentro da correlação serial temos a **Autocorrelação** que utilizada como uma forma de medida da correlação entre os valores da série temporal e seus próprios valores defasados (As defasagens na autocorrelação são os intervalos de tempo entre os valores da série temporal que são usados para calcular a correlação. Por exemplo, uma defasagem de 1 significa que a autocorrelação é calculada entre o valor atual da série temporal e seu valor no período anterior. Uma defasagem de 2 significa que a autocorrelação é calculada entre o valor atual da série temporal e seu valor dois períodos antes.).


```python
# exemplo de cálculo
med_mov = pd.Series(np.arange(10))
med_mov.rolling(3).mean()
```

##**Métricas para avaliar modelos de Série Temporal**

Antes de entrar nos algoritimos de predição é interessante dá uma rápida passada pelas métricas usadas para avaliar os modelos, aqui será abordada 4 métricas. Para medir um modelo (seja ele qual for), vamos medir e analisar os erros que ele apresenta, ou seja, vamos comparar Y e Ŷ (Y real e Y previsto, respectivamente) e dar atenção à esses resíduos.

###**Erro Médio Absoluto - MAE**

O erro médio absoluto, MAE (da sigla em inglês Mean Absolute Error), é calculado a partir da média dos erros absolutos, ou seja, utilizamos o módulo de cada erro para evitar a subestimação, isso porque, o valor é menos afetado por pontos especialmente extremos (outliers).

Utilizamos essa medida em séries temporais, pois há casos em que o erro negativo pode zerar o positivo ou dar uma ideia de que o modelo é preciso. Mas aqui, medimos apenas a distância do valor real, independente de ser acima ou abaixo.

<center><img width="25%" src="https://raw.githubusercontent.com/Yago-Pacheco/apresentacoes_pr/main/mae.jpg">

```python
# exemplo
from sklearn.metrics import mean_absolute_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
mean_absolute_error(y_true, y_pred)
```

###**Erro Quadrático Médio - MSE**

O erro quadrático médio, MSE (da sigla em inglês Mean Squared Error), é comumente usado para verificar a acurácia de modelos e dá um maior peso aos maiores erros, já que, ao ser calculado, cada erro é elevado ao quadrado individualmente e, após isso, a média desses erros quadráticos é calculada.

Por conta do expoente ao quadrado que o erro assume, essa métrica é bastante sensível a outliers (valores discrepantes) e, caso tenha muitos erros significativos em sua análise, essa métrica poderá ser extrapolada.

<center><img width="25%" src="https://raw.githubusercontent.com/Yago-Pacheco/apresentacoes_pr/main/mse.jpg">

```python
# exemplo
from sklearn.metrics import mean_squared_error

y_true = [[0.5, 1],[-1, 1],[7, -6]]
y_pred = [[0, 2],[-1, 2],[8, -5]]

mean_squared_error(y_true, y_pred, squared=True)
```

###**Raiz Quadrada do Erro Médio - RMSE**

A raiz quadrada do erro médio, RMSE (da sigla em inglês Root Mean Squared Error) é apenas a raiz quadrada do MSE, onde o erro retorna à unidade de medida do modelo (no MSE, a unidade de medida é quadrática).

É frequentemente usada em séries temporais por ser mais sensível a erros maiores devido ao processo de quadratura que a produziu.

<center><img width="25%" src="https://raw.githubusercontent.com/Yago-Pacheco/apresentacoes_pr/main/rmse.jpg">

```python
# exemplo
from sklearn.metrics import mean_squared_error

y_true = [[0.5, 1],[-1, 1],[7, -6]]
y_pred = [[0, 2],[-1, 2],[8, -5]]

mean_squared_error(y_true, y_pred, squared=False)
```

###**Média Percentual Absoluta do Erro - MAPE**

Esta é outra métrica interessante para usar, geralmente usada em relatórios de gerenciamento, porque o erro é medido como uma porcentagem e assim, é possível fazer comparações entre erros percentuais do modelo entre produtos.

<center><img width="25%" src="https://raw.githubusercontent.com/Yago-Pacheco/apresentacoes_pr/main/mape.jpg">

```python
# exemplo
from sklearn.metrics import mean_absolute_percentage_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

mean_absolute_percentage_error(y_true, y_pred)
```

Profeta é um procedimento para previsão de dados de séries temporais com base em um modelo aditivo onde tendências não lineares são ajustadas à sazonalidade anual, semanal e diária, além dos efeitos dos feriados. Alguns conceitos legais para explicar sobre o `Prophet` é que foi desenvolvido pela META e por ser um modelo aditivo, o que significa que é composto de três componentes: tendência, sazonalidade e feriados.

* **Tendência:** A tendência é o componente que captura o movimento de longo prazo da série temporal. O Prophet usa um modelo de tendência linear ou logística para estimar a tendência.



* **Sazonalidade:** A sazonalidade é o componente que captura os padrões recorrentes na série temporal. O Prophet usa um modelo de sazonalidade de Fourier para estimar a sazonalidade.


* **Feriados:** O componente de feriados captura o impacto dos feriados na série temporal. O Prophet usa um modelo de efeitos de feriados para estimar o impacto dos feriados.

O Prophet é um modelo flexível que pode ser usado para uma variedade de aplicações. Ele é especialmente adequado para séries temporais que apresentam tendências, sazonalidade e feriados. Tem grande uso em:

* Previsão de vendas
* Previsão de tráfego
* Previsão de preços
* Previsão de eventos climáticos

### **ARIMA**

O modelo ARIMA é um modelo de séries temporais que é uma combinação de três modelos: autorregressivo (AR), integrado (I) e de média móvel (MA).

* **Autorregressivo (AR)**: O modelo AR assume que os valores da série temporal são influenciados por seus próprios valores passados. Por exemplo, um modelo AR(1) assume que o valor atual da série temporal é influenciado pelo valor do período anterior.

* **Integrado (I)**: O modelo I é usado para tornar a série temporal estacionária. Uma série temporal é estacionária se sua média, variância e autocorrelação não mudarem ao longo do tempo.

* **Média móvel (MA)**:O modelo MA assume que os valores da série temporal são influenciados por erros passados. Por exemplo, um modelo MA(1) assume que o valor atual da série temporal é influenciado pelo erro do período anterior.


A notação ARIMA(p, d, q) representa o número de termos no modelo AR, o grau de diferenciação e o número de termos no modelo MA. Por exemplo, um modelo ARIMA(1, 1, 1) tem um termo AR de primeira ordem, um termo MA de primeira ordem e não é diferenciado.

Os modelos ARIMA são uma classe de modelos de séries temporais muito populares. Eles são capazes de capturar uma variedade de padrões em séries temporais, incluindo tendências, sazonalidade e autocorrelação.

Aqui estão alguns exemplos de como os modelos ARIMA podem ser usados:

* Previsão de vendas
* Previsão de tráfego
* Previsão de preços
* Previsão de eventos climáticos

Por curiosidade aqui tem uma rápida explicação dos termos `p, d, q`.

* **P**: O número de termos autorregressivos (AR) no modelo representa o número de valores passados da série temporal que são usados para prever o valor atual. Por exemplo, um modelo ARIMA(1, 1, 1) tem um termo AR de primeira ordem, o que significa que o valor atual da série temporal é previsto usando o valor do período anterior.


* **D**: O grau de diferenciação da série temporal é o número de vezes que a série temporal é diferenciada para torná-la estacionária. Uma série temporal é estacionária se sua média, variância e autocorrelação não mudarem ao longo do tempo.

* **Q**: O número de termos de média móvel (MA) no modelo representa o número de erros passados da série temporal que são usados para prever o valor atual. Por exemplo, um modelo ARIMA(1, 1, 1) tem um termo MA de primeira ordem, o que significa que o valor atual da série temporal é previsto usando o erro do período anterior.



### **SARIMA**

O modelo SARIMA é uma extensão do modelo ARIMA que inclui um componente sazonal. O modelo SARIMA é usado para prever séries temporais que apresentam padrões sazonais. A sazonalidade é a tendência de uma série temporal de se repetir em intervalos regulares, geralmente em intervalos de um dia, semana, mês, trimestre ou ano.

A notação SARIMA(p, d, q)(P, D, Q)m representa os seguintes componentes do modelo:

* p, d, q: Os mesmos componentes do modelo ARIMA
* P, D, Q: Os mesmos componentes do modelo ARIMA, mas aplicados aos dados sazonais
* m: O número de períodos sazonais nos dados

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

df = pd.read_csv('https://raw.githubusercontent.com/Yago-Pacheco/apresentacoes_pr/main/AirPassengers.csv', parse_dates=[0])

df = df.set_index('Month') # passando a coluna para índice
df = df['#Passengers']
```

```python
df.tail()
```

```python
t_treino = df[df.index<'1959-11-01'].index.values
t_teste = df[df.index>='1959-11-01'].index.values
X_treino = df[df.index<'1959-11-01'].values
X_teste = df[df.index>='1959-11-01'].values

modelo = SARIMAX(X_treino, order=(2,1,1), seasonal_order=(0,1,1,12))
resultado = modelo.fit()

X_pred = resultado.predict(start=len(X_treino), end=len(X_treino)+len(X_teste)-1)
```

```python
plt.plot(t_treino,X_treino,label='Treino')
plt.plot(t_teste,X_teste,label='Teste')
plt.plot(t_teste,X_pred,label='Predito')
plt.xticks(rotation=45)
plt.legend()
```

```python
# calculando os erros
RMSE = mean_squared_error(X_teste,X_pred, squared=False)
print('RMSE =  {:0.2f}'.format(RMSE))
```
